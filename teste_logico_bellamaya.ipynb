```python
import requests
import os
from dotenv import load_dotenv

# Carrega suas configurações
load_dotenv()

WAHA_BASE_URL = os.getenv("WAHA_BASE_URL", "http://localhost:3000")
WAHA_API_KEY = os.getenv("WAHA_API_KEY", "")

def listar_grupos():
    url = f"{WAHA_BASE_URL}/api/groups"
    headers = {
        "X-Api-Key": WAHA_API_KEY
    }
    params = {
        "session": "default" # Usando a sessão padrão conforme suas memórias
    }

    try:
        print(f"📡 Consultando grupos em: {url}...")
        response = requests.get(url, headers=headers, params=params)

        if response.status_code == 200:
            grupos = response.json()
            print(f"\n✅ Encontrei {len(grupos)} grupos:\n")
            print(f"{'NOME DO GRUPO':<30} | {'ID (COPIE ESTE)'}")
            print("-" * 60)

            for grupo in grupos:
                nome = grupo.get('name', 'Sem Nome')
                id_grupo = grupo.get('id', 'Sem ID')
                # Filtra apenas grupos reais (terminam em @g.us)
                if "@g.us" in id_grupo:
                    print(f"{nome:<30} | {id_grupo}")
        else:
            print(f"❌ Erro ao buscar grupos: {response.text}")

    except Exception as e:
        print(f"❌ Erro de conexão: {e}")

# Executa a busca
listar_grupos()
```
